```python
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import ConfusionMatrixDisplay
from google.colab import files

#uploaded = files.upload()

df = pd.read_csv('credit_data.csv')
df = df[df.age >0].dropna()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['income', 'age', 'loan']], df.default, test_size=0.2, random_state=42)
x_test
```

```python
from sklearn.metrics import accuracy_score

acc = []
for k in range(1, 20, 2):
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(x_train, y_train)
  acc.append((k, accuracy_score(y_test, knn.predict(x_test))))
```

```python
from sklearn.metrics import confusion_matrix
knn2 = KNeighborsClassifier(n_neighbors=5)
knn2.fit(x_train, y_train)
cm2 = confusion_matrix(y_test, knn2.predict(x_test))
ConfusionMatrixDisplay(cm2).plot()
```

```python
knn = KNeighborsClassifier(n_neighbors=45)
knn.fit(x_train, y_train)
cm = confusion_matrix(y_test, knn.predict(x_test))

ConfusionMatrixDisplay(cm).plot()
```
